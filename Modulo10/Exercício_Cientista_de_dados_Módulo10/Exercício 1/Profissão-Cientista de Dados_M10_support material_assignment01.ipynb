````python
#Importação das Bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
#Lendo CSV e Amostragem de Dados.

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
#Variáveis não importantes neste exercício.

df.drop(['Unnamed: 0', 'index', 'mau'],axis=1, inplace=True)
````

````python
#Analisando Quantidade de Linhas, Colunas, Tipos de Dados e NA.

df.info()
````

````python
#Retirando as NA e transformando objects e bool em int.

df.dropna(inplace=True)
df['sexo'] = df['sexo'].map({'F': 1, 'M': 0})
df['posse_de_veiculo'] = df['posse_de_veiculo'].map({True: 1, False: 0})
df['posse_de_imovel'] = df['posse_de_imovel'].map({True: 1, False: 0})
df.info()
````

Uma instituição financeira quer conhecer melhor o perfil de renda de seus novos clientes para diversos fins, por exemplo, melhor dimensionar o limite de cartões de crédito dos novos clientes, sem necessariamente solicitar olerites ou documentações que impactem na experiência do seu cliente.

Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

As variáveis são intuitivas - note que há uma variável 'index' que é um identificador do cliente e que em geral o ```read_csv``` do pandas coloca também uma variável sequencial.

Estes dados estão no arquivo ```previsao_de_renda.csv```. Carregue-o em um *dataframe*.

# 2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

* As duas variáveis mais correlacionadas com renda são **'idade'** e **'tempo_emprego'**.

````python
plt.rc('figure',figsize=(15,15))

fig, axes = plt.subplots(2,2)

ax1 = sns.scatterplot(ax=axes[0,0], x='idade', y='renda', data = df, hue='sexo')
ax1.set_title("RENDA x IDADE (SCATTERPLOT)")

ax2 = sns.scatterplot(ax=axes[0,1], x='tempo_emprego', y='renda', data = df, hue='sexo')
ax2.set_title("RENDA x TEMPO_EMPREGO (SCATTERPLOT)")

# Pode-se visualizar atraves do (*regplot*) uma correlação positiva entre essas duas variaveis.

ax3 = sns.regplot(ax=axes[1,0], x='idade', y='renda', data = df, line_kws={'color':'red'})
ax3.set_title("RENDA x IDADE (REGPLOT)")

ax4 = sns.regplot(ax=axes[1,1], x='tempo_emprego', y='renda', data = df, line_kws={'color':'red'})
ax4.set_title("RENDA x TEMPO_EMPREGO (REGPLOT)")

plt.savefig('Análise1.jpg')
````

# 3) Na sua aopinião, há outliers na variável renda?

* Sim, há vários outliers na variável **renda**, devido as variaveis **'idade'** e **'tempo_emprego'** apesar de ter uma maior correlação, ainda sim, é muito baixa.
  Sendo:
  * **'idade'**: 0.15
  * **'tempo_emprego'**: 0.39.

# 4) Calcule o logaritmo da variável renda e repita a sequência de análise

* Pela nova sequência de análise, percebe-se que a linha de correlação ficou mais centralizada, porém ainda contém vários outliers.

````python
df_log = df.copy()

df_log["renda_log"] = np.log(df_log['renda'])

df_log.corr(numeric_only=True)
````

````python
plt.rc('figure',figsize=(15,15))

fig, axes = plt.subplots(2,2)

plt.subplots_adjust(wspace=0.3, hspace=0.3)

ax1 = sns.scatterplot(ax=axes[0,0], x='idade', y='renda_log', data = df_log, hue='sexo')
ax1.set_title("RENDA x IDADE (SCATTERPLOT)")

ax2 = sns.scatterplot(ax=axes[0,1], x='tempo_emprego', y='renda_log', data = df_log, hue='sexo')
ax2.set_title("RENDA x TEMPO_EMPREGO (SCATTERPLOT)")

# Pode-se visualizar atraves do (*regplot*) uma correlação positiva entre essas duas variaveis.

ax3 = sns.regplot(ax=axes[1,0], x='idade', y='renda_log', data = df_log, line_kws={'color':'red'})
ax3.set_title("RENDA x IDADE (REGPLOT)")

ax4 = sns.regplot(ax=axes[1,1], x='tempo_emprego', y='renda_log', data = df_log, line_kws={'color':'red'})
ax4.set_title("RENDA x TEMPO_EMPREGO (REGPLOT)")

plt.savefig('Análise2.jpg')
````

````python
sns.clustermap(data=df_log[['renda_log', 'idade', 'tempo_emprego']].corr(numeric_only=True),
               center=0,
               annot=True,
               cmap='RdYlGn',
               figsize=(4,4))

plt.savefig('Clustermap2.jpg')
````
