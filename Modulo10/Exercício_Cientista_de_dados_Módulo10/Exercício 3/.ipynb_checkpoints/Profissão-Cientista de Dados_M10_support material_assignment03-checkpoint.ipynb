````python
# Importando Bibliotecas.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
````

````python
# Lendo CSV e Amostragem de Dados.

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# Analisando numero de linhas, colunas, tipos de dados e NA's.

df.info()
````

````python
# Retirando as variáveis que não serão necessárias para essa análise.

df.drop(['Unnamed: 0', 'index', 'qtd_filhos', 'tempo_emprego', 'idade', 'qt_pessoas_residencia'], axis=1, inplace=True)
````

````python
# Mapping coluna 'sexo'.
df['sexo'] = df['sexo'].map({'F': 'Feminino', 'M': 'Masculino'})

# Formatando 'data_ref' para o tipo 'datetime'.
df['data_ref'] = pd.to_datetime(df['data_ref'], format='%Y/%m/%d')

# Mapping coluna 'data_ref' para String em formato: mm-YYYY (ex.: 01-2015).
df['data_ref'] = df['data_ref'].map(lambda ts: ts.strftime('%m-%Y'))

# Nova df.info() para analisar as informações que ficaram.
df.info()
````

# Módulo 10 tarefa 03
Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

### 1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

**Resposta Exercício 1**

O gráfico se mantem constante ao longo do tempo em praticamente todas as variáveis qualitativas, não houve nenhuma mudança significativa, porém na variável **'educacao'** percebemos uma queda da população analisada com Ensino **'Secundário'** e um aumento da população analisada com **'Superior Completo'**.

````python
# Função para fazer os gráficos do exercício 1.

# column: Coluna a ser analisada; ax: axes para adicionar nos subplots; titulo: Título do gráfico.
def countplot_ex1(column,ax,titulo=''):
    
    # Função para visualizar o gráfico.
    sns.countplot(ax=ax,
                  x='data_ref',
                  hue = column,
                  data=df)
    
    # Título do Gráfico.
    ax.set_title(titulo) 
    
    # Definindo labels do eixo x e rotacionando em 90°.
    ax.set_xticklabels(ax.get_xticklabels(),
                       rotation=90)
    ax.set_ylabel('Contagem')
    ax.set_xlabel('Data')
    
    # Posicionando as legendas.
    ax.legend(bbox_to_anchor=(1.05, 1),
              loc='best',
              borderaxespad=0.0,
              title='Legenda:')
````

````python
# Define o tamanho da figura.
plt.rc('figure',figsize=(10,25))

# Cria subplots com 8 linhas e 1 coluna.
fig, axes = plt.subplots(8,1)

# Chamando função countplot_ex1.
ax1 = countplot_ex1('sexo',axes[0],'COUNTPLOT DA VARIÁVEL SEXO')
ax2 = countplot_ex1('posse_de_veiculo',axes[1],'COUNTPLOT DA VARIÁVEL POSSE_DE_VEICULO')
ax3 = countplot_ex1('posse_de_imovel',axes[2],'COUNTPLOT DA VARIÁVEL POSSE_DE_IMOVEL')
ax4 = countplot_ex1('tipo_renda',axes[3],'COUNTPLOT DA VARIÁVEL TIPO_RENDA')
ax5 = countplot_ex1('educacao',axes[4],'COUNTPLOT DA VARIÁVEL EDUCACAO')
ax6 = countplot_ex1('estado_civil',axes[5],'COUNTPLOT DA VARIÁVEL ESTADO_CIVIL')
ax7 = countplot_ex1('tipo_residencia',axes[6],'COUNTPLOT DA VARIÁVEL TIPO_RESIDENCIA')
ax8 = countplot_ex1('mau',axes[7],'COUNTPLOT DA VARIÁVEL MAU')

# Ajuste automatico do espaçamento entre os gráficos.
plt.tight_layout()
````

### 2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

**Resposta Exercício 2**

* Análise de cada variável Qualitativa em relação as variáveis **'renda' x 'data_ref'**:
    
    **'sexo'**: O gráfico se manteve estável ao longo da variável tempo.
    
    **'posse_de_veiculo'**: O gráfico teve um alteração relevante nos meses **08/2015** e **09/2015**.
    
    **'posse_de_imovel'**: O gráfico apresenta uma alteração relevante nos meses **10/2015** e **12-2015**.
    
    **'tipo_renda'**: O gráfico não possue informações suficientes em relação aos **Bolsistas**, o restante aprenseta dados estáveis.
    
    **'educacao'**: O gráfico apresenta instabilidade para população com **'Pós graduação'**, o restante apresenta dados estáveis.
    
    **'estado_civil'**: O gráfico apresenta dados estáveis, com excessão do mês **03-2015** para a população denominada **'União'** que houve um certo aumento de renda. A população denominada **'Separado'** apresenta uma instabilidade maior ao longo do tempo.
    
    **'tipo_residencia'**: O gráfico apresenta uma instabilidade maior na população denominada **'Comunitário'**, o restante aprenseta dados estáveis.
    
    **'mau'**: O gráfico apresenta uma maior renda dos **'maus'** pagadores nos meses **06-2015** e **01-2016**.
    
<div class="alert alert-block alert-info">
<b>Obs:</b> Para uma análise mais detalhada, deve-se olhar os gráficos mais de perto, além de utilizar outras formas de análise.
</div>

````python
# Função para fazer os gráficos do exercício 2.

# column: Coluna a ser analisada; ax: axes para adicionar nos subplots; titulo: Título do gráfico.
def pointplot_ex2(column,ax,titulo=''):
    
    # Função para visualizar o gráfico.
    sns.pointplot(ax=ax,
                  x='data_ref',
                  y='renda',
                  hue = column,
                  data=df,
                  dodge=True,
                  errorbar=('ci', 95))
    
    # Título do Gráfico.
    ax.set_title(titulo)
    
    # Definindo labels do eixo x e rotacionando em 90°.
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    ax.set_xlabel('Data')
    
    # Posicionando as legendas.
    ax.legend(bbox_to_anchor=(1.05, 1), loc='best', borderaxespad=0.0, title='Legenda:') 
````

````python
# Define o tamanho da figura.
plt.rc('figure',figsize=(10,30))

# Cria subplots com 8 linhas e 1 coluna.
fig, axes = plt.subplots(8,1)

# Chamando função pointplot_ex2.
ax1 = pointplot_ex2('sexo',axes[0],'POINTPLOT DA VARIÁVEL SEXO')
ax2 = pointplot_ex2('posse_de_veiculo',axes[1],'POINTPLOT DA VARIÁVEL POSSE_DE_VEICULO')
ax3 = pointplot_ex2('posse_de_imovel',axes[2],'POINTPLOT DA VARIÁVEL POSSE_DE_IMOVEL')
ax4 = pointplot_ex2('tipo_renda',axes[3],'POINTPLOT DA VARIÁVEL TIPO_RENDA')
ax5 = pointplot_ex2('educacao',axes[4],'POINTPLOT DA VARIÁVEL EDUCACAO')
ax6 = pointplot_ex2('estado_civil',axes[5],'POINTPLOT DA VARIÁVEL ESTADO_CIVIL')
ax7 = pointplot_ex2('tipo_residencia',axes[6],'POINTPLOT DA VARIÁVEL TIPO_RESIDENCIA')
ax8 = pointplot_ex2('mau',axes[7],'POINTPLOT DA VARIÁVEL MAU')

# Ajuste automatico do espaçamento entre os gráficos.
plt.tight_layout()
````
