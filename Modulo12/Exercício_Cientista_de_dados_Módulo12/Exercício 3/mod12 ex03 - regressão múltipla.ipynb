### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, time e net_bill```.  
2. Remova as variáveis não significantes. **(variável 'sex' removida)**
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
#### Resposta 4.
- **'net_bill':** O R² **(0.335)** é bom , **'time[T.Dinner]'** ruim, porém ainda dá para manter.
- **'np.log(net_bill)':** O R² **(0.295)** diminui, porém o Pvalue de **'time[T.Dinner]'** melhora e o de **'smoker[T.No]'** piora.
- **'np.power(net_bill,2)':** O R² **(0.305)** é um pouco melhor que o do **'np.log(net_bill)'**, o Pvalue de **'time[T.Dinner]'** e o de **'smoker[T.No]'** melhoram. Considero esta a melhor opção devido os Pvalues melhorarem e o R² ser OK.

````python
# 1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em tip,
#    explicada por sex, smoker, time e net_bill.

y1, X1 = patsy.dmatrices('tip ~ smoker + time + net_bill', data=tips)
X1
````

````python
# 2. Remova as variáveis não significantes.

reg1 = sm.OLS(y1, X1).fit()

reg1.summary()
````

````python
# 3. Observe o gráfico de resíduos em função de net_bill

tips['res_y1'] = reg1.resid
sns.scatterplot(x = 'net_bill', y = 'res_y1', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# 4. teste transformar net_bill no log e um polinômio. Escolha o melhor modelo.
#    np.log(net_bill)

y2, X2 = patsy.dmatrices('tip ~ smoker + time + np.log(net_bill)', data=tips)
X2
````

````python
# 2. Remova as variáveis não significantes.
# 4. np.log(net_bill)

reg2 = sm.OLS(y2, X2).fit()

reg2.summary()
````

````python
# 3. Observe o gráfico de resíduos em função de net_bill
# 4. np.log(net_bill)

tips['res_y2'] = reg2.resid
sns.scatterplot(x = np.log(tips['net_bill']), y = 'res_y2', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# 4. teste transformar net_bill no log e um polinômio. Escolha o melhor modelo.
#    np.power(net_bill,2)

y3, X3 = patsy.dmatrices('tip ~ smoker + time + np.power(net_bill,2)', data=tips)
X3
````

````python
# 2. Remova as variáveis não significantes.
# 4. np.power(net_bill,2)

reg3 = sm.OLS(y3,X3).fit()

reg3.summary()
````

````python
# 3. Observe o gráfico de resíduos em função de net_bill
# 4. np.power(net_bill,2)

tips['res_y3'] = reg3.resid
sns.scatterplot(x = np.log(tips['net_bill']), y = 'res_y3', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````
