## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o percentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
#Análise inicial
dfVestibular = pd.DataFrame()
dfVestibular['Matematica'] = df_mat
dfVestibular['Portugues'] = df_por
dfVestibular['Geografia'] = df_geo
dfVestibular['Ingles'] = df_ing
dfVestibular['Historia'] = df_his
dfVestibular['Fisica'] = df_fis
dfVestibular['Quimica'] = df_qui

dfVestibular
````

````python
# 1) (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.

print(f'Matemática:\n{(dfVestibular.Matematica/24*100).head().round(2)}\n')
print(f'Português:\n{(dfVestibular.Portugues/18*100).head().round(2)}\n')
print(f'Geografia:\n{(dfVestibular.Geografia/8*100).head().round(2)}\n')
print(f'Inglês:\n{(dfVestibular.Ingles/8*100).head().round(2)}\n')
print(f'História:\n{(dfVestibular.Historia/8*100).head().round(2)}\n')
print(f'Física:\n{(dfVestibular.Fisica/12*100).head().round(2)}\n')
print(f'Química:\n{(dfVestibular.Quimica/12*100).head().round(2)}\n')
````

````python
# 2) (operações entre *DataFrames) Calcule o total de acertos de cada aluno.
dfVestibular['Soma_Total'] = dfVestibular.T['Matematica':'Quimica'].sum()
dfVestibular
````

````python
# 3) Calcule o percentual geral de cada aluno.
dfVestibular['Percentural_geral'] = round(dfVestibular['Soma_Total']/90*100,2)
dfVestibular
````

````python
# 4) Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?
dfVestibular.loc[dfVestibular['Soma_Total'] > 45, 'Soma_Total'].count()
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
df_eliminados = pd.DataFrame(dfVestibular.loc[(dfVestibular['Matematica'] == 0) | 
                                              (dfVestibular['Fisica'] == 0) |
                                              (dfVestibular['Quimica'] == 0)])
# 1) Alunos desqualificados
print(f"Alunos desqualificados:\n{df_eliminados[['Matematica','Fisica','Quimica']]}\n")

# 2) Soma dos Alunos eliminados
print(f"Eliminados:\n{df_eliminados[['Soma_Total']].count()}\n")

# 3) Média desses Alunos em História e Geografia
print(f"Média(História e Geografia):\n{df_eliminados[['Historia', 'Geografia']].mean().round(2)}\n")

# 4) DataFrame dos Alunos que passaram (>45 acertos, além de acertos em Matemática, Física e Química)
df_aprovados = dfVestibular.loc[(dfVestibular['Soma_Total'] > 45) & (dfVestibular['Matematica'] > 0) & (dfVestibular['Fisica'] > 0) & (dfVestibular['Quimica'] > 0)]
print(f"Alunos Aprovados:\n{df_aprovados[['Matematica', 'Fisica', 'Quimica', 'Soma_Total', 'Percentural_geral']]}\n")
print(f"Quantidade de Alunos aprovados: {df_aprovados.Soma_Total.count()}")
````

````python
# 3) Quantos pacientes únicos tomaram a primeira dose? OBS: Há um caractere especial neste campo.
primeira_dose = df['paciente_id'].loc[df['vacina_descricao_dose'] == df['vacina_descricao_dose'].unique()[0]].nunique()
print(f"*{primeira_dose}* tomaram a primeira dose.")
````

````python
# 4) Quantos pacientes com menos de 18 anos foram vacinados?
df_menor = df[df["paciente_idade"]<18]
print(f"{df_menor['paciente_id'].nunique()} pacientes que foram vacinados tem menos de 18 anos.")
````
