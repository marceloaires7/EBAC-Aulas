### Qual o maior vencedor de jogos?

```python
flamengo = ((brasileirao['Mandante'] == 'Flamengo')|(brasileirao['Visitante'] == 'Flamengo'))&(brasileirao['Data'] >= '2019-01-01')&(brasileirao['Data'] <= '2020-01-01')
print(f"{brasileirao[flamengo]['Vencedor'].value_counts()}\n\n")
Vitorias = brasileirao[flamengo&(brasileirao['Vencedor']=='Flamengo')]['Vencedor'].count()
Derrotas = brasileirao[flamengo&(brasileirao['Vencedor']!='Flamengo')&(brasileirao['Vencedor']!='-')]['Vencedor'].count()
Empates = brasileirao[flamengo&(brasileirao['Vencedor']=='-')]['Vencedor'].count()
print(f"Vitórias: {Vitorias}\nDerrotas: {Derrotas}\nEmpates: {Empates}\n")
```

```python
brasileirao[flamengo]
```

### Qual a tabela de jogos vencidos por clube na temporada de 2020? 
(periodo temp = '2020-01-01' a '2021-02-26')

```python
cond = (brasileirao['Data'] >= '2020-01-01')& (brasileirao['Data'] <= '2021-02-26')
cond
```

```python
brasileirao[cond]['Vencedor'].value_counts()
```

## MeiliSearch
- https://github.com/meilisearch/MeiliSearch
- https://raw.githubusercontent.com/meilisearch/MeiliSearch/main/datasets/movies/movies.json

```python
movies = pd.read_json('https://raw.githubusercontent.com/meilisearch/meilisearch-python/14179ef7678550c6f53d04f1780dfdcbfb6e3a82/datasets/small_movies.json')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies['ano'] = movies['release_date'].astype('datetime64[s]').dt.strftime('%Y')
```

```python
movies['mes'] = movies['release_date'].astype('datetime64[s]').dt.strftime('%m')
```

```python
movies['dia'] = movies['release_date'].astype('datetime64[s]').dt.strftime('%d')
```

```python
movies.head()
```

### Quais são os tipos de dados das colunas?

```python
movies.dtypes
```

### Qual genêro predominante no conjunto de dados?

```python
movies.genre.explode().value_counts()
```

### Qual o ano de lançamento predominante da base??

```python
movies['ano'].value_counts()
```
