```python
#Importações das Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#%matplotlib notebook
```

```python
#Lendo o CSV e Amostra dos Dados.
df = pd.read_csv('sinasc_RO_2019.csv')
df.head()
```

```python
#Quantidade de linhas, colunas, dados, dtypes de cada dado e nome das colunas.
df.info()
```

```python
#Detalhamento dos Dados.
df.describe()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
#BOXPLOT - Idade da Mãe
sns.boxplot(y='IDADEMAE',
            data=df,
            color='yellow',
            flierprops={'marker': 'o', 'markerfacecolor': 'red'})

plt.title('BOXPLOT - Idade da Mãe') #Definindo título do gráfico.
plt.ylabel('Idade da Mãe') #Nomeando eixo Y.

plt.savefig('BOXPLOT - Idade da Mãe.jpg')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close('all') #Fechando os gráficos anteriores.

#BOXPLOT - Peso do Bebê
sns.boxplot(data=df,
            y='PESO',
            color='green',
            flierprops={'marker': 'o', 'markerfacecolor': 'red'})

plt.title('BOXPLOT - Peso do Bebê')#Definindo título do gráfico.
plt.ylabel('Peso')#Nomeando eixo Y.

plt.savefig('BOXPLOT - Peso do Bebê.jpg')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('all') #Fechando os gráficos anteriores.

#HISTPLOT - Idade da Mãe
sns.histplot(data=df, #Data.
             x='IDADEMAE', #Coluna.
             bins=41, #Número de intervalos no gráfico.
             color='yellow', #Cor das barras.
             kde=True).lines[0].set_color('red') #Linha KDE e cor da linha.

plt.title('HISTPLOT - Idade da Mãe') #Título do gráfico.
plt.xlabel('Idade da Mãe') #Nomeando Eixo X.
plt.ylabel('Quantidade') #Nomeando Eixo Y.

plt.savefig('HISTPLOT - Idade da Mãe.jpg')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.close('all') #Fechando os gráficos anteriores.

#HISTPLOT - Idade da Mãe
sns.histplot(data=df, #Data.
             x='PESO', #Coluna.
             bins=41, #Número de intervalos no gráfico.
             color='green', #Cor das barras.
             kde=True).lines[0].set_color('red') #Linha KDE e cor da linha.

plt.title('HISTPLOT - Peso do Bebê') #Título do gráfico.
plt.xlabel('Peso do Bebê') #Nomeando Eixo X.         
plt.ylabel('Quantidade') #Nomeando Eixo Y.

plt.savefig('HISTPLOT - Peso do Bebê.jpg')
plt.show()
```
