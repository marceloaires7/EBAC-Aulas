```python
def Legenda(grafico=None, rotacao=None, titulo=None): #Fórmula para colocar legenda no gráfico.
    grafico
    
    ax = plt.gca()  # Obter o eixo atual.
    
    for p in ax.patches: #for para criar a legenda.
        ax.annotate(f'{round(p.get_height(),2)}', #Texto a ser escrito no gráfico.
                    xy=(p.get_x() + p.get_width() / 2, p.get_height()), #Localização do Texto.
                    ha='center', #Alinhamento Horizontal.
                    va='bottom', #Alinhamento Vertical.
                    rotation=rotacao) #Rotação do texto.
    
    plt.title(titulo) #Título do gráfico.
```

```python
#Lendo CSV e Amostra de Dados.
df = pd.read_csv('sinasc_RO_2019.csv')
df.head()
```

```python
#Informações dos dados (Colunas, Quantidade de Linhas, Tipo de Dados)
df.info()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5'] <= 3,['APGAR5_CAT']] = 'asfixia grave' #Adicionando 'asfixia grave'.
df.loc[(df['APGAR5'] > 3)&(df['APGAR5'] <= 5),['APGAR5_CAT']] = 'asfixia moderada' #Adicionando 'asfixia moderada'.
df.loc[(df['APGAR5'] > 5)&(df['APGAR5'] <= 7),['APGAR5_CAT']] = 'asfixia leve' #Adicionando 'asfixia leve'.
df.loc[df['APGAR5'] >= 8,['APGAR5_CAT']] = 'normal' #Adicionando 'normal'.

df['APGAR5_CAT'].value_counts() #Quantidade adicionada de cada item.
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
#Utilização da Função Legenda em cima do gráfico que criamos.
Legenda(sns.catplot(data = df, #DataFrame utilizado para analise.
                    x = 'GESTACAO', #Dado do Eixo x.
                    y = 'APGAR5', #Dado do Eixo y.
                    kind = 'bar', #Tipo de gráfico.
                    hue = 'SEXO', #Dado extra para mostrar separado.
                    palette = ['green', 'yellow'],
                    errorbar = None), #Não mostrar Intervalo de Confiança.
        70,
        'Distribuição de APGAR5 por sexo para cada grupo GESTACAO\n') #Título do Gráfico.
plt.tick_params(axis='x', rotation=70) #Rotação da legenda no Eixo x.
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
#Motando tabela cruzada com a frequência que 'APGAR5' aparece em cada 'GESTACAO'
freq = pd.crosstab(df['GESTACAO'], df['APGAR5'])
freq
```

####  5) Da tabela cruzada do item anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
#Calculo do percentual de cada item da Tabela.
freq_pct = freq.div(freq.sum(axis=1),axis=0)
freq_pct
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
plt.close('all') #Fechando os gráficos anteriores.

Legenda(grafico = (freq_pct*100).plot.bar(width=0.7, figsize=(10,10)),
        rotacao = 90,
        titulo = 'Gráfico de Barras da Tabela Cruzada Acima')
#O gráfico não fica bem visível, mas dá para notar uma queda nos 'APGAR5' para 'GESTACAO' abaixo das 37 Semanas.
#Pelo gráfico de barras com a tabela acima não dá para concluir, acredito que um outro tipo de gráfico seja melhor para análise.
```

```python
plt.close('all') #Fechando os gráficos anteriores.

#Com um BARPLOT entre 'GESTACAO' e 'APGAR5', podemos analisar melhor que com menos de 37 semanas o valor do APGAR5 começa a cair.
Legenda(sns.barplot(data=df, x='GESTACAO', y='APGAR5', errorbar=None),
        0,
        'BARPLOT - "GESTACAO" x "APGAR5"')
plt.tick_params(axis='x', rotation=30) #Rotação do label x.
```
